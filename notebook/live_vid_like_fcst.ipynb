# 3. Predicting like counts in live sessions

```python
import pandas as pd
import numpy as np

# Work on a copy to avoid side effects
df = df_live.copy()

# ========== FEATURE ENGINEERING ==========

# 1. Engagement Score = (likes + shares + comments) / userCount
df["engagement_score"] = (
    df["stats.diggCount"].fillna(0)
    + df["stats.shareCount"].fillna(0)
    + df["stats.commentCount"].fillna(0)
) / df["liveRoomInfo.liveRoomStats.userCount"].replace(0, np.nan)

# 2. Duration bins (if you have duration column)
if "video.duration" in df.columns:
    df["duration_bins"] = pd.cut(df["video.duration"], bins=[0, 60, 300, 1800, np.inf],
                                 labels=["Short (<1 min)", "Medium (1-5 min)", "Long (5-30 min)", "Very Long (>30 min)"])
else:
    df["duration_bins"] = "Unknown"

# 3. Time of Day (if createTime is UNIX timestamp in seconds)
df["create_hour"] = pd.to_datetime(df["createTime"], unit='s', errors='coerce').dt.hour
df["time_of_day"] = pd.cut(df["create_hour"],
                           bins=[-1, 6, 12, 18, 24],
                           labels=["Late Night", "Morning", "Afternoon", "Evening"])

# 4. is_verified_creator
df["is_verified_creator"] = df["author.verified"].fillna(False).astype(int)

# 5. Use CategoryType directly
df["category_type"] = df["CategoryType"]

# ========== SELECT RELEVANT COLUMNS ==========
selected_columns = [
    "liveRoomInfo.likeCount",
    "liveRoomInfo.liveRoomStats.userCount",
    "likes_per_user",
    "engagement_score",
    "duration_bins",
    "time_of_day",
    "is_verified_creator",
    "category_type",
    "stats.diggCount",
    "stats.shareCount",
    "stats.commentCount",
    "create_hour"
]

df_model = df[selected_columns].copy()

# Optional: Drop rows with missing userCount or likeCount
df_model.dropna(subset=["liveRoomInfo.likeCount", "liveRoomInfo.liveRoomStats.userCount"], inplace=True)

# Final check
print("✅ Feature engineering complete. Sample:")
print(df_model.head())
```

```python
# Replace NaNs with 0 for interaction metrics
interaction_cols = ["stats.diggCount", "stats.shareCount", "stats.commentCount"]
df_model[interaction_cols] = df_model[interaction_cols].fillna(0)

# Recompute engagement score
df_model["engagement_score"] = (
    df_model["stats.diggCount"] +
    df_model["stats.shareCount"] +
    df_model["stats.commentCount"]
) / df_model["liveRoomInfo.liveRoomStats.userCount"].replace(0, np.nan)
```

# 4. Simpler Machine Learning Models 

## 4a. Ridge/ Lasso Regression

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# ---- Select features and target ----
target = 'liveRoomInfo.likeCount'
features = df_model.drop(columns=[target])
y = df_model[target]

# ---- Identify numeric and categorical columns ----
numeric_cols = features.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = features.select_dtypes(include=['object', 'category']).columns.tolist()

# ---- Preprocessing ----

# One-hot encode categorical columns
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
X_cat = encoder.fit_transform(features[categorical_cols])
X_cat_df = pd.DataFrame(X_cat, columns=encoder.get_feature_names_out(categorical_cols))

# Scale numeric columns
scaler = StandardScaler()
X_num = scaler.fit_transform(features[numeric_cols])
X_num_df = pd.DataFrame(X_num, columns=numeric_cols)

# Combine processed features
X_processed = pd.concat([X_num_df.reset_index(drop=True), X_cat_df.reset_index(drop=True)], axis=1)

# ---- Train-test split ----
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.25, random_state=42)

# ---- Train Ridge and Lasso ----
ridge = Ridge(alpha=1.0)
lasso = Lasso(alpha=0.1)

ridge.fit(X_train, y_train)
lasso.fit(X_train, y_train)

# ---- Predict and evaluate ----
ridge_pred = ridge.predict(X_test)
lasso_pred = lasso.predict(X_test)

ridge_mae = mean_absolute_error(y_test, ridge_pred)
lasso_mae = mean_absolute_error(y_test, lasso_pred)

# ---- Print results ----
print(f"Ridge Regression MAE: {ridge_mae:.2f}")
print(f"Lasso Regression MAE: {lasso_mae:.2f}")
```

## 4b. Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor


# ---- Separate target and features ----
target = 'liveRoomInfo.likeCount'
features = df_model.drop(columns=[target])
y = df_model[target]

# ---- Separate numeric and categorical features ----
numeric_cols = features.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = features.select_dtypes(include=['object', 'category']).columns.tolist()

# ---- Preprocess ----
# One-hot encode categoricals
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
X_cat = encoder.fit_transform(features[categorical_cols])
X_cat_df = pd.DataFrame(X_cat, columns=encoder.get_feature_names_out(categorical_cols))

# Standard scale numeric features
scaler = StandardScaler()
X_num = scaler.fit_transform(features[numeric_cols])
X_num_df = pd.DataFrame(X_num, columns=numeric_cols)

# Combine
X_processed = pd.concat([X_num_df.reset_index(drop=True), X_cat_df.reset_index(drop=True)], axis=1)

# ---- Train-Test Split ----
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.25, random_state=42)

# ---- Train Random Forest Regressor ----
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# ---- Predict and Evaluate ----
y_pred = rf_model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)

# ---- Results ----
print(f"🌲 Random Forest MAE: {mae:.2f}")
```

Comparing simpler machine learning model and deep learning neural network, deep learning neural network model performs much better, showing a result of at least 4 times better than that of simpler machine learning models.
