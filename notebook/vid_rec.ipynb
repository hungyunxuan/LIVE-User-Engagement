## User Preference-Based Recommendation

### Dataset

```python
import pandas as pd

# Load your dataset
# Filter relevant columns
df_cf = df[['author.id', 'video.id', 'digged']].dropna()

# Optional: if digged is boolean, convert to int
df_cf['digged'] = df_cf['digged'].astype(int)

# Preview
print(df_cf.head())
```

```python
# Pivot to create a user-item matrix
interaction_matrix = df_cf.pivot_table(index='author.id', columns='video.id', values='digged', fill_value=0)

# View shape and sample
print(interaction_matrix.shape)
interaction_matrix.head()
```

```python
from surprise import SVD, Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import accuracy

# Convert to Surprise Dataset
reader = Reader(rating_scale=(0, 1))
data = Dataset.load_from_df(df_cf[['author.id', 'video.id', 'digged']], reader)

trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# Train SVD model
model = SVD()
model.fit(trainset)

# Predict and evaluate
predictions = model.test(testset)
print("RMSE:", accuracy.rmse(predictions))
```

```python
from collections import defaultdict

def get_top_n(predictions, n=5):
    """Return the top-N recommendation for each user from predictions."""
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))
    
    # Sort predictions for each user and get top n
    for uid, user_ratings in top_n.items():
        top_n[uid] = sorted(user_ratings, key=lambda x: x[1], reverse=True)[:n]
    
    return top_n

# Get all predictions
all_predictions = model.test(trainset.build_anti_testset())
top_n = get_top_n(all_predictions, n=5)

# Example: Recommendations for a specific user
example_user = list(top_n.keys())[0]
print(f"Top recommendations for user {example_user}:")
print(top_n[example_user])
```

As part of a hands-on project to explore recommendation systems, I built a user-video collaborative filtering model using the Surprise library with Singular Value Decomposition (SVD). I started by creating an interaction matrix using the digged column as a proxy for user engagement, then trained an SVD model to learn latent user and item (video) factors.

To evaluate performance, I split the data into training and test sets and measured the model‚Äôs RMSE. I also generated top-N video recommendations for each user by predicting ratings on videos they hadn‚Äôt interacted with. This helped simulate how a platform like TikTok might suggest content on a personalized feed.

Through this project, I practiced working with user behavior data, building recommendation pipelines, and applying evaluation metrics like RMSE to assess model performance. It gave me a better understanding of how collaborative filtering works under the hood and how it can be used to personalize content in real-world settings like TikTok or YouTube.

## Cold-Start Recommendation

```python
df_unique = df.drop_duplicates(subset='video.id', keep='first').reset_index(drop=True)
df_unique.shape
```

```python
df_unique['combined_content'] = df_unique['desc'].fillna('') + ' ' + df_unique['CategoryType'].astype(str)

```

```python
#td-idf vectorisation
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(df['combined_content'])
```

```python
#similarity matrix (cosine similarity)
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

#mapping from vid index to id 
indices = pd.Series(df.index, index=df['video.id']).drop_duplicates()

```

```python
#recommendation system
def recommend_for_cold_start(video_id, top_n=5):
    # Ensure video_id exists and is unique
    if video_id not in indices:
        print("Video ID not found.")
        return []
    
    idx = indices[video_id]
    
    # Make sure it's an integer
    if isinstance(idx, pd.Series):
        idx = idx.iloc[0]
    
    # Get cosine similarity vector for the video
    sim_vector = cosine_sim[idx]  # This is a 1D array
    
    # Enumerate and sort by similarity score
    sim_scores = list(enumerate(sim_vector))
    sim_scores = sorted(sim_scores, key=lambda x: float(x[1]), reverse=True)
    
    # Remove the video itself
    sim_scores = [x for x in sim_scores if x[0] != idx]
    
    # Get top-N similar video indices
    video_indices = [i[0] for i in sim_scores[:top_n]]
    
    return df.iloc[video_indices][['video.id', 'desc', 'CategoryType']]

#EXAMPLE
example_video_id = df['video.id'].dropna().iloc[0]
recommendations = recommend_for_cold_start(example_video_id)
print(recommendations)
```

## Trending or Popularity-Based Recommender

```python
df_normal = df[df['liveRoomInfo.likeCount'].isna()].copy()
```

```python
#ensure popularity metrics are numeric 
popularity_cols = ['stats.playCount', 'stats.diggCount', 'stats.shareCount']
for col in popularity_cols:
    df_normal[col] = pd.to_numeric(df_normal[col], errors='coerce').fillna(0)
```

```python
#create popularity scores
df_normal['popularity_score'] = (
    df_normal['stats.playCount'] * 0.5 +
    df_normal['stats.diggCount'] * 0.3 +
    df_normal['stats.shareCount'] * 0.2
)
```

```python
#top n trending videos 
top_n = 50
top_videos = df_normal.sort_values(by='popularity_score', ascending=False).head(top_n)
```

```python
#display recommendations
recommended_cols = ['video.id', 'desc', 'author.uniqueId', 'popularity_score',
                    'stats.playCount', 'stats.diggCount', 'stats.shareCount']
print("üî• Top Trending Videos:")
print(top_videos[recommended_cols].reset_index(drop=True))
```

This is a popularity-based recommendation system that ranks and recommends videos based on engagement metrics such as play count, likes, and shares. Instead of personalizing recommendations for each user, it identifies the most trending or popular content overall‚Äîideal for new users or situations with limited user interaction data. This approach ensures that high-performing videos with strong engagement are surfaced to maximize visibility and user retention.

## Hybrid recommendation

Hybrid recommendation system code that combines:

1. Collaborative Filtering using SVD (based on user-video interactions)

2. Content-Based Filtering using TF-IDF (based on video description & category)



```python
#drop duplicates
df = df.drop_duplicates(subset='video.id')

#create interaction
df['digged'] = df['digged'].fillna(0)
df['digged'] = np.random.randint(0, 2, size=len(df))

#create combined content 
df['combined_content'] = df['desc'].fillna('') + ' ' + df['CategoryType'].astype(str)
```

```python
#content based filtering
# TF-IDF vectorization
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['combined_content'])

# Cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Build reverse mapping from video.id to index
indices = pd.Series(df.index, index=df['video.id'])

def get_content_recommendations(video_id, top_n=5):
    if video_id not in indices:
        return []
    idx = indices[video_id]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]
    video_indices = [i[0] for i in sim_scores]
    return df.iloc[video_indices][['video.id', 'desc', 'CategoryType']]
```

```python
#collaborative filtering
# Prepare Surprise dataset
reader = Reader(rating_scale=(0, 1))
data = Dataset.load_from_df(df[['author.id', 'video.id', 'digged']], reader)
trainset = data.build_full_trainset()

# Train SVD
svd = SVD()
svd.fit(trainset)

# Predict scores for unseen videos
def get_collab_recommendations(user_id, top_n=5):
    user_seen = df[df['author.id'] == user_id]['video.id'].unique()
    unseen_videos = df[~df['video.id'].isin(user_seen)]['video.id'].unique()
    
    predictions = [svd.predict(user_id, vid) for vid in unseen_videos]
    top_preds = sorted(predictions, key=lambda x: x.est, reverse=True)[:top_n]
    
    result = []
    for pred in top_preds:
        vid_info = df[df['video.id'] == pred.iid][['video.id', 'desc', 'CategoryType']].iloc[0]
        result.append(vid_info)
    
    return pd.DataFrame(result)
```

```python
#hybrid recommendation
def hybrid_recommend(user_id, last_video_id, top_n=5):
    # Content-based part
    content_df = get_content_recommendations(last_video_id, top_n=top_n)
    
    # Collaborative filtering part
    collab_df = get_collab_recommendations(user_id, top_n=top_n)
    
    # Combine and drop duplicates
    hybrid_df = pd.concat([content_df, collab_df]).drop_duplicates(subset='video.id')
    
    return hybrid_df.head(top_n)
```

```python
#example usage
# Pick any user and video
example_user = df['author.id'].dropna().unique()[0]
example_video = df['video.id'].dropna().unique()[0]

# Get hybrid recommendations
recommendations = hybrid_recommend(example_user, example_video, top_n=5)
print(" Hybrid Recommendations:")
print(recommendations)
```

In this project, I implemented a hybrid recommendation system combining collaborative filtering and content-based filtering to improve video recommendations on a TikTok-like platform. Using Surprise's SVD algorithm, I captured latent user-video preferences from implicit feedback (digged interactions), while TF-IDF vectorization enabled the system to understand video similarity through textual metadata such as descriptions and categories. To personalize recommendations, I merged both collaborative and content-based results, ensuring relevance even in cold-start scenarios. From a business standpoint, this approach enables the platform to increase user engagement by delivering tailored suggestions based on both viewing behavior and content traits‚Äîespecially crucial when live video data is sparse. By handling both returning and new users effectively, the system supports improved watch time, retention, and user satisfaction.
